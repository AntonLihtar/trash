```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

```python
#Линейные графики
x = list(range(2015, 2026))
y = [89, 89.5, 89.5, 90, 89.5, 86.5, 89, 95, 100, 107, 115 ]



plt.plot(x, y, marker='o')  # добавим точки для наглядности
plt.title("Простой линейный график")
plt.xlabel("Года")
plt.ylabel("Вес в кг")
plt.xticks(x)  # <-- показывает все года


plt.show()
plt.close('all')
```

```python
#Диаграммы (столбчатые, круговые и др.)
plt.bar(['A', 'B', 'C'], [10, 20, 15])
plt.show()

plt.pie([40, 35, 25], labels=['Apple', 'Banana', 'Cherry'])
plt.show()
```

```python
#Гистограммы
import numpy as np
data = np.random.randn(1000)  # 1000 случайных чисел
plt.hist(data, bins=30)
plt.show()
```

```python
#Множественные графики
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.bar(['A', 'B', 'C'], [3, 7, 5])

plt.show()
```
