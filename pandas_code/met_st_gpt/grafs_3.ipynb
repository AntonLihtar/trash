```python
import matplotlib.pyplot as plt

# Данные
days = [1, 2, 3, 4, 5]
temperatures = [5, 7, 6, 8, 7]      # для линейного графика
sales = [10, 15, 7, 12, 9]           # для круговой диаграммы
data_points = [5, 7, 6, 8, 7]        # для гистограммы
```

```python
# 1️⃣ Линейный график
plt.figure()
plt.plot(days, temperatures)
plt.title('Line plot - Температура по дням')
plt.xlabel('День')
plt.ylabel('Температура')
plt.show()

# 2️⃣ Гистограмма
plt.figure()
plt.hist(data_points, bins=5)
plt.title('Histogram - Распределение температур')
plt.xlabel('Температура')
plt.ylabel('Частота')
plt.show()

# 3️⃣ Круговая диаграмма
categories = ['A', 'B', 'C', 'D', 'E']
plt.figure()
plt.pie(sales, labels=categories)
plt.title('Pie chart - Продажи по категориям')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
days = [1, 2, 3, 4, 5]
temperatures = [5, 7, 6, 8, 7]      # Линейный
data_points = [5, 7, 6, 8, 7]       # Гистограмма
sales = [10, 15, 7, 12, 9]          # Круговая диаграмма
categories = ['A', 'B', 'C', 'D', 'E']

# Создаём фигуру с сеткой 2x2
fig = plt.figure(figsize=(10, 8))

# 1️⃣ Гистограмма - верхний левый угол
ax1 = plt.subplot2grid((2,2), (0,0))
ax1.hist(data_points, bins=5)
ax1.set_title('Histogram')

# 2️⃣ Круговая диаграмма - верхний правый угол
ax2 = plt.subplot2grid((2,2), (0,1))
ax2.pie(sales, labels=categories)
ax2.set_title('Pie chart')

# 3️⃣ Линейный график - нижняя строка (оба столбца)
ax3 = plt.subplot2grid((2,2), (1,0), colspan=2)
ax3.plot(days, temperatures)
ax3.set_title('Line plot')
ax3.set_xlabel('День')
ax3.set_ylabel('Температура')

plt.tight_layout()  # чтобы графики не накладывались
plt.show()
```
