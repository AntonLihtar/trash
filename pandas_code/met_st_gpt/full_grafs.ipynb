```python
import matplotlib.pyplot as plt

# Данные
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]
data = [1,2,2,3,3,3,4,4,4,4,5,5,6,6,7]
```

```python
# 1️⃣ Линейный график
plt.figure()
plt.plot(x, y)
plt.title('Line plot')
plt.show()
```

```python
# 2️⃣ Точечный график
plt.figure()
plt.scatter(x, y)
plt.title('Scatter plot')
plt.show()
```

```python
# 3️⃣ Столбчатый график
plt.figure()
plt.bar(categories, values)
plt.title('Bar chart')
plt.show()
```

```python
# 4️⃣ Горизонтальный столбчатый график
plt.figure()
plt.barh(categories, values)
plt.title('Horizontal Bar chart')
plt.show()
```

```python
# 5️⃣ Гистограмма
plt.figure()
plt.hist(data, bins=6)
plt.title('Histogram')
plt.show()
```

```python
# 6️⃣ Круговая диаграмма
plt.figure()
plt.pie(values, labels=categories)
plt.title('Pie chart')
plt.show()
```

```python
# 7️⃣ Fill between (область под линией)
y2 = [3, 5, 4, 6, 5]
plt.figure()
plt.plot(x, y, label='y')
plt.plot(x, y2, label='y2')
plt.fill_between(x, y, y2)
plt.title('Fill between')
plt.show()
```

```python
# 8️⃣ Логарифмический график
x_log = [1, 10, 100, 1000, 10000]
y_log = [1, 10, 100, 1000, 10000]
plt.figure()
plt.plot(x_log, y_log)
plt.xscale('log')
plt.yscale('log')
plt.title('Logarithmic plot')
plt.show()
```

```python
# 9️⃣ Box plot
plt.figure()
plt.boxplot(data)
plt.title('Box plot')
plt.show()
```
