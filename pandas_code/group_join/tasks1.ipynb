```python
import pandas as pd

employees = pd.DataFrame([
    ['IT', 'Developer', 120000, 3],
    ['IT', 'Developer', 130000, 5],
    ['IT', 'Analyst', 100000, 2],
    ['IT', 'Analyst', 105000, 3],
    ['HR', 'Recruiter', 70000, 4],
    ['HR', 'Manager', 90000, 6],
    ['HR', 'Recruiter', 75000, 2],
    ['Sales', 'Salesperson', 80000, 1],
    ['Sales', 'Manager', 120000, 7],
    ['Sales', 'Salesperson', 82000, 2],
    ['Marketing', 'Specialist', 65000, 3],
    ['Marketing', 'Manager', 95000, 5],
    ['Marketing', 'Specialist', 67000, 2],
    ['Finance', 'Accountant', 85000, 4],
    ['Finance', 'Manager', 110000, 6],
    ['Finance', 'Accountant', 87000, 3],
    ['IT', 'Developer', 140000, 6],
    ['Sales', 'Salesperson', 81000, 3],
    ['HR', 'Recruiter', 72000, 3],
    ['Marketing', 'Specialist', 68000, 4],
], columns=['department', 'position', 'salary', 'experience_years'])

print(employees)
```

```python
# Вывести среднюю зарплату в каждом отделе.
mean_salary_from_dep = employees.groupby(by=['department'])['salary'].agg('mean')
mean_salary_from_dep
```

```python
# Вывести среднюю зарплату по каждой комбинации (department, position).
mean_salary_from_dep_position = employees.groupby(by=['department', 'position'])['salary'].mean()
mean_salary_from_dep_position.unstack()
# mean_salary_from_dep_position
```

```python
# Добавить столбец с разницей между зарплатой сотрудника и средней по отделу.
mean_sal = employees.groupby('department')['salary'].mean()
# print(mean_sal)
employees['difference_sal'] = employees['salary'] -  employees['department'].map(mean_sal)
employees
```

```python
mean_sal = employees.groupby('department', as_index=False)['salary'].mean()
mean_sal.rename(columns={'salary': 'salary_mean'}, inplace=True)


employees.merge(mean_sal, on='department', how='left')

employees
```
