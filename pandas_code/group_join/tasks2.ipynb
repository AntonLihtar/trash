```python
import pandas as pd

sales = pd.DataFrame([
    ['Jan', 'Alice', 'Laptop', 120000],
    ['Jan', 'Bob', 'Laptop', 95000],
    ['Jan', 'Charlie', 'Phone', 78000],
    ['Feb', 'Alice', 'Laptop', 110000],
    ['Feb', 'Bob', 'Laptop', 122000],
    ['Feb', 'Charlie', 'Phone', 81000],
    ['Mar', 'Alice', 'Laptop', 125000],
    ['Mar', 'Bob', 'Laptop', 98000],
    ['Mar', 'Charlie', 'Phone', 85000],
    ['Apr', 'Alice', 'Laptop', 140000],
    ['Apr', 'Bob', 'Laptop', 110000],
    ['Apr', 'Charlie', 'Phone', 87000],
    ['May', 'Alice', 'Laptop', 135000],
    ['May', 'Bob', 'Laptop', 108000],
    ['May', 'Charlie', 'Phone', 190000],
    ['Jun', 'Alice', 'Laptop', 145000],
    ['Jun', 'Bob', 'Laptop', 112000],
    ['Jun', 'Charlie', 'Phone', 92000],
    ['Jul', 'Alice', 'Laptop', 100000],
    ['Jul', 'Bob', 'Laptop', 115000],
], columns=['month', 'manager', 'product', 'revenue'])

sales
```

```python
# Для каждого месяца найти менеджера с максимальной выручкой.
sales.groupby('month', sort=False).agg({'revenue': 'max'})
```

```python
# Вывести таблицу с колонками month | best_manager | revenue_max.
revenue_max_df = sales.groupby('month', sort=False)['revenue'].max().reset_index()
revenue_max_df.rename(columns={'revenue': 'revenue_max'}, inplace=True)
# print(revenue_max_df)
n_sales = pd.merge(left=sales, right=revenue_max_df, left_on='month', right_on='month', how='left')
# print(n_sales)

n_sales = n_sales[n_sales['revenue'] == n_sales['revenue_max']]
n_sales['best_manager'] = n_sales['manager']
res_sal = n_sales[['month', 'best_manager', 'revenue_max']]
res_sal
```

```python
# Проверить, сколько раз каждый менеджер был “лучшим” за всё время.
best_count = res_sal['best_manager'].value_counts()
best_count
```
